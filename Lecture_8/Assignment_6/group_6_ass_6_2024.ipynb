### 3. Use Google API to find the driving time (best guess) from all the group members' address and all the LIMA BBVA offices.

```python
#Se crean dos listas, una para las direcciones y otra para sus correspondientes coordenadas 
direcciones = ["Av. Las Américas 371, Independencia", "Cristóbal de Peralta norte 950, Surco", "Calle Guadalajara 272, Ate", "Jiron Centenario 1365, Breña", "Condominios Tomas Valle, Los Olivos"]

coordenadas = []

#Se crea un loop para cada dirección y asi extraer su latitud y longitud
for direccion in direcciones:
    dir_mar = gmaps.geocode(direccion, region='pe')
    lati_dir_mar = dir_mar[0]['geometry']['location']['lat']
    long_dir_mar = dir_mar[0]['geometry']['location']['lng']
    coord_mar = str(lati_dir_mar) + ',' + str(long_dir_mar)
    
    # Se divide el string en latitud y longitud
    latitud_str, longitud_str = coord_mar.split(',')

    # Se convierte el string a float
    latitud = float(latitud_str)
    longitud = float(longitud_str)

    # Crear una tupla con latitud y longitud
    coord_mar_tupla = (latitud, longitud)
    
    coordenadas.append(coord_mar_tupla)

for i, coord_house in enumerate(coordenadas):
    print(f"Las coordenadas de la dirección {direcciones[i]} son: {coord_house}")
```

```python
# Función para obtener el tiempo de conducción desde las coordenadas de los integrantes hasta todas las oficinas de BBVA en LIMA
def obtener_tiempos_conduccion(coordenadas_inicio, coordenadas_final):
    # Generando la clave API
    gmaps = googlemaps.Client(key='AIzaSyDQ8OA89GFnZvFhG03W2NFpRIDM1ad_Ff4')

    # Lista para almacenar los tiempos de conducción
    tiempos_conduccion = []

    # Itera sobre las coordenadas de las oficinas de BBVA
    for coordenada_bbva in coordenadas_final:
        # Realiza la solicitud de tiempo de conducción a Google Maps
        directions_result = gmaps.directions(
            origin=','.join(map(str, coordenadas_inicio)),  # Convertir la tupla de inicio a una cadena
            destination=','.join(map(str, coordenada_bbva)),  # Convertir la tupla BBVA a una cadena
            mode="driving",  # Modo de transporte: driving
            departure_time="now",  # Hora de salida: now
            traffic_model="best_guess"  # Mejor estimación del tráfico
        )
        # Se obtiene el tiempo de conducción estimado
        if directions_result and 'legs' in directions_result[0]:
            tiempo_conduccion = directions_result[0]['legs'][0]['duration']['value']
        else:
            tiempo_conduccion = 'N/A'  # para cuando no se pueda obtener el tiempo de conducción

        # Agrega el tiempo de conducción a la lista
        tiempos_conduccion.append(tiempo_conduccion)

    return tiempos_conduccion
```

### 4. Finally, you have to give a report which offices are the most closest and furthest to every group member's address.

```python
# Función para obtener la oficina más cercana y más lejana desde las coordenadas de los integrantes hasta todas las oficinas de BBVA en LIMA
def obtener_oficinas_cercana_lejana(coordenadas_inicio, lista_oficinas, coordenadas_final):
    # Generando la clave API
    gmaps = googlemaps.Client(key='AIzaSyDQ8OA89GFnZvFhG03W2NFpRIDM1ad_Ff4')

    # Diccionario para almacenar los tiempos de conducción
    tiempos_conduccion = {}

    # Itera sobre las coordenadas de las oficinas de BBVA
    for i, coordenada_bbva in enumerate(coordenadas_final):
        # Realiza la solicitud de tiempo de conducción a Google Maps
        directions_result = gmaps.directions(
            origin=','.join(map(str, coordenadas_inicio)),  # Convertir la tupla de inicio a una cadena
            destination=','.join(map(str, coordenada_bbva)),  # Convertir la tupla BBVA a una cadena
            mode="driving",  # Modo de transporte: driving
            departure_time="now",  # Hora de salida: now
            traffic_model="best_guess"  # Mejor estimación del tráfico
        )
        # Se obtiene el tiempo de conducción estimado
        if directions_result and 'legs' in directions_result[0]:
            tiempo_conduccion = directions_result[0]['legs'][0]['duration']['value']
        else:
            tiempo_conduccion = float('inf')  # para cuando no se pueda obtener el tiempo de conducción

        # Agrega el tiempo de conducción al diccionario con la oficina como clave
        tiempos_conduccion[lista_oficinas[i]] = tiempo_conduccion

    # Encuentra la oficina más cercana y más lejana
    oficina_cercana = min(tiempos_conduccion, key=tiempos_conduccion.get)
    oficina_lejana = max(tiempos_conduccion, key=tiempos_conduccion.get)

    return oficina_cercana, oficina_lejana
```
